## 🛠️ Creating a Simple Tool

```python
def add(a: float, b: float) -> float:
    """Adds two numbers together.
    
    Args:
        a: The first number.
        b: The second number.
    
    Returns:
        The sum of a and b.
    """
    return a + b

def multiply(a: float, b: float) -> float:
    """Multiplies two numbers together.
    
    Args:
        a: The first number.
        b: The second number.
    
    Returns:
        The product of a and b.
    """
    return a * b
```

## 🤖 Creating an Agent with Tools

```python
from camel.agents import ChatAgent
from camel.toolkits import FunctionTool
from camel.types import ModelType

# Create tools
add_tool = FunctionTool(add)
multiply_tool = FunctionTool(multiply)

# Create agent with tools
agent = ChatAgent(
    system_message="You are a helpful math assistant. Use the provided tools to solve math problems.",
    model=ModelType.GPT_4O_MINI,
    tools=[add_tool, multiply_tool]
)
```

## 💬 Using the Agent

```python
# Ask a math question
response = agent.step("What is 5 + 3?")
print(response.msgs[0].content)
```

```python
# Ask a more complex question
response = agent.step("What is (5 + 3) * 2?")
print(response.msgs[0].content)
```
