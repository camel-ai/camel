## Overview

This cookbook demonstrates how to build a **Data Analysis Agent** using CAMEL's built-in data toolkits. We combine multiple data sources into a single intelligent agent that can:

- 🌍 Query real-world statistical data using `DataCommonsToolkit`
- 📈 Retrieve financial market data using `OpenBBToolkit`
- 🤖 Analyse and answer natural language questions using `ChatAgent`

This is a practical end-to-end example of how CAMEL's multi-toolkit agent approach enables powerful data-driven insights with minimal code.

## 3. Import Libraries

```python
from camel.agents import ChatAgent
from camel.messages import BaseMessage
from camel.models import ModelFactory
from camel.toolkits import DataCommonsToolkit
from camel.types import ModelPlatformType, ModelType

print("Libraries imported successfully! 🐫")
```

## 4. Create the Model

```python
model = ModelFactory.create(
    model_platform=ModelPlatformType.OPENAI,
    model_type=ModelType.GPT_4O_MINI,
    model_config_dict={"temperature": 0.0},
)

print(f"Model created: {ModelType.GPT_4O_MINI}")
```

## 5. Part 1 — Statistical Data Analysis with DataCommonsToolkit

`DataCommonsToolkit` provides access to Google's Data Commons knowledge graph, which contains statistical data on population, health, economics, environment, and more — no API key required!

```python
# Initialize DataCommonsToolkit
data_commons_toolkit = DataCommonsToolkit()
data_tools = data_commons_toolkit.get_tools()

print(f"DataCommonsToolkit loaded with {len(data_tools)} tools:")
for tool in data_tools:
    print(f"  - {tool.func.__name__}")
```

```python
# Create a Data Analysis Agent with DataCommonsToolkit
sys_msg = BaseMessage.make_assistant_message(
    role_name="Data Analyst",
    content=(
        "You are an expert data analyst with access to the Data Commons "
        "knowledge graph. Use the available tools to retrieve and analyse "
        "real-world statistical data. Always provide clear insights and "
        "explain your findings in plain language."
    ),
)

data_agent = ChatAgent(
    system_message=sys_msg,
    model=model,
    tools=data_tools,
)

print("Data Analysis Agent ready! 🤖")
```

### Query 1: Population Statistics

```python
# Query population data
query1 = BaseMessage.make_user_message(
    role_name="User",
    content=(
        "What is the total population of California, USA? "
        "Use the Data Commons tools to find this information."
    ),
)

response1 = data_agent.step(query1)
print("Query: What is the total population of California?")
print("-" * 60)
print(response1.msgs[0].content)
```

### Query 2: Time Series Statistical Data

```python
# Reset agent memory for a fresh query
data_agent.reset()

query2 = BaseMessage.make_user_message(
    role_name="User",
    content=(
        "Retrieve the CO2 emissions time series data for the United States "
        "and identify the trend. What does this tell us about climate change?"
    ),
)

response2 = data_agent.step(query2)
print("Query: CO2 emissions trend for the United States")
print("-" * 60)
print(response2.msgs[0].content)
```

### Query 3: Comparative Analysis Across Regions

```python
# Reset agent memory for a fresh query
data_agent.reset()

query3 = BaseMessage.make_user_message(
    role_name="User",
    content=(
        "Compare the median household income across the top 5 most populous "
        "US states. Which state has the highest income and what might explain "
        "this difference?"
    ),
)

response3 = data_agent.step(query3)
print("Query: Median household income across top 5 US states")
print("-" * 60)
print(response3.msgs[0].content)
```

## 6. Part 2 — Multi-Toolkit Data Analysis Agent

Now let's combine **multiple toolkits** into a single powerful agent that can answer complex questions spanning different data sources.

```python
from camel.toolkits import SearchToolkit

# Combine DataCommonsToolkit + SearchToolkit for richer analysis
search_toolkit = SearchToolkit()
search_tools = search_toolkit.get_tools()

# Combine all tools
all_tools = data_tools + search_tools

# Create a more powerful multi-toolkit agent
multi_sys_msg = BaseMessage.make_assistant_message(
    role_name="Senior Data Analyst",
    content=(
        "You are a senior data analyst with access to multiple data sources. "
        "You can query statistical databases via Data Commons and search the web "
        "for current information. Combine insights from multiple sources to "
        "provide comprehensive, well-rounded analysis. Always cite your data "
        "sources and quantify your findings where possible."
    ),
)

multi_agent = ChatAgent(
    system_message=multi_sys_msg,
    model=model,
    tools=all_tools,
)

print(f"Multi-toolkit agent created with {len(all_tools)} tools! 🚀")
```

```python
# Complex cross-source analysis query
complex_query = BaseMessage.make_user_message(
    role_name="User",
    content=(
        "Analyse the relationship between urbanisation rates and GDP per capita "
        "for major economies. Use statistical data to support your analysis and "
        "provide actionable insights for policymakers."
    ),
)

complex_response = multi_agent.step(complex_query)
print("Query: Urbanisation vs GDP analysis")
print("-" * 60)
print(complex_response.msgs[0].content)
```
